```python
# Định nghĩa hàm is_group_member
def is_group_member(group_data, n):
    # Duyệt qua từng phần tử (value) trong danh sách group_data.
    for value in group_data:
        # Kiểm tra xem giá trị hiện tại có bằng với số n hay không.
        if n == value: 
            return True  # trả về True
    return False  # Trả về False nếu vòng lặp kết thúc mà không tìm thấy giá trị nào khớp.

# Gọi hàm với các bộ dữ liệu khác nhau và in kết quả. 
print(is_group_member([1, 5, 8, 3], 3))  
print(is_group_member([5, 8, 3], -1))  
```

```python
def khong_de_quy(n):
    if n <= 1: return n
    a, b = 0, 1
  
    for _ in range(2, n + 1):
        tmp = a + b  # Cộng tổng
        a = b        # Đổi chỗ
        b = tmp      # Cập nhật số mới nhất   
    return b

# Test
n = 10
print(f"Số Fibonacci thứ {n} (không đệ quy) là: {khong_de_quy(n)}")
```

```python
def tim_lon_thu_nhi(mang):
    tap_hop_so = list(set(mang))  # Loại bỏ trùng lặp bằng set
    tap_hop_so.sort() # Sắp xếp
    
    # Nếu có ít nhất 2 phần tử, số áp chót là số lớn thứ nhì
    if len(tap_hop_so) < 2:
        return "Không có số lớn thứ nhì"   
    return tap_hop_so[-2]

A = [10, 5, 10, 8, 3, 8, 9]
print("Số lớn thứ nhì là:", tim_lon_thu_nhi(A))
```

```python
def tim_lon_thu_nhi(mang_goc):
    mang = mang_goc.copy() # Copy để không làm hỏng mảng gốc
    max_val = max(mang)
    
    # Xóa tất cả các phần tử bằng giá trị lớn nhất
    while max_val in mang:
        mang.remove(max_val)
    
    if not mang:
        return "Không có"
    return max(mang)

print("Số lớn thứ nhì là:", tim_lon_thu_nhi([10, 10, 5, 2]))
```

```python
def tim_so_xuat_hien_n_lan(mang, n_lan):
    dem = {}
    for x in mang:
        dem[x] = dem.get(x, 0) + 1
    
    # Lọc ra các số xuất hiện đúng n_lan
    ket_qua = [so for so, lan in dem.items() if lan == n_lan]
    return ket_qua

A = [1, 2, 2, 3, 3, 3, 4, 1, 5]
n = 2
print(f"Các số xuất hiện đúng {n} lần là:", tim_so_xuat_hien_n_lan(A, n))
```

```python
from collections import Counter

def tim_so_xuat_hien_n_lan(mang, n_lan):
    thong_ke = Counter(mang)
    return [so for so, lan in thong_ke.items() if lan == n_lan]

A = [4, 4, 5, 6, 6, 7, 4]
print(f"Các số xuất hiện đúng {n} lần là:", tim_so_xuat_hien_n_lan(A, 2)) 
```

```python
def tim_so_xuat_hien_nhieu_nhat(mang):
    if not mang: return []
    dem = {}
    for x in mang:
        dem[x] = dem.get(x, 0) + 1
    
    max_lan = max(dem.values())  # Tìm số lần xuất hiện lớn nhất
    
    # Lấy tất cả các số có số lần xuất hiện bằng max_lan
    ket_qua = [so for so, lan in dem.items() if lan == max_lan]
    return ket_qua, max_lan

mang_so = [1, 2, 3, 1, 2, 1, 4, 2]
so_nhieu_nhat, so_lan = tim_so_xuat_hien_nhieu_nhat(mang_so)
print(f"Các số xuất hiện nhiều nhất ({so_lan} lần): {so_nhieu_nhat}")
```

```python
# Định nghĩa hàm "difference" nhận vào một tham số là số nguyên "n"
def difference(n):
    # Kiểm tra nếu n nhỏ hơn hoặc bằng 17 (nếu thỏa thì trả về hiệu số giữa 17 và n)
    if n <= 17:
        return 17 - n
    else:
        # Nếu n lớn hơn 17, trả về hiệu số giữa n và 17 rồi nhân đôi kết quả
        return (n - 17) * 2

# Gọi hàm "difference" với tham số là 22 và in kết quả
print(difference(22)) # Vì 22 > 17, kết quả sẽ là (22 - 17) * 2 = 5 * 2 = 10

# Gọi hàm "difference" với tham số là 14 và in kết quả
print(difference(14)) # Vì 14 <= 17, kết quả sẽ là 17 - 14 = 3
```

```python
# Định nghĩa hàm "near_thousand" nhận vào một tham số là số nguyên "n"
def near_thousand(n):
    	# Kiểm tra xem giá trị tuyệt đối của (1000 - n) có nhỏ hơn hoặc bằng 100 không
# HOẶC kiểm tra xem giá trị tuyệt đối của (2000 - n) có nhỏ hơn hoặc bằng 100 không     
    return ((abs(1000 - n) <= 100) or (abs(2000 - n) <= 100))
# Gọi hàm với các giá trị khác nhau và in kết quả (True hoặc False)
print(near_thousand(1000))

print(near_thousand(900))

print(near_thousand(800))

print(near_thousand(2200))
```

```python
# Định nghĩa hàm sum_thrice
def sum_thrice(x, y, z):
    sum = x + y + z     # Tính tổng 3 số x,y,z
    # Kiểm tra xem cả ba số x, y và z có bằng nhau hay không
    if x == y == z:
        sum = sum * 3 #nhân tổng đó lên 3 lần

    # Trả về kết quả tổng
    return sum

# Gọi hàm với các giá trị khác nhau và in kết quả 
print(sum_thrice(1, 2, 3))
print(sum_thrice(3, 3, 3))
```

```python
# Định nghĩa hàm "new_string" 
def new_string(text):
    # Kiểm tra nếu độ dài của chuỗi "text" lớn hơn hoặc bằng 2 và hai ký tự đầu tiên của chuỗi đó là "Is"
    if len(text) >= 2 and text[:2] == "Is":
       # Nếu điều kiện trên đúng, trả về chuỗi gốc mà không thay đổi gì
        return text
    else:
        # Nếu không, thêm chữ "Is" vào trước chuỗi đó rồi trả về
        return "Is" + text

# Gọi hàm với đối số là "Array" và in kết quả
print(new_string("Array"))

# Gọi hàm với đối số là "IsEmpty" và in kết quả
print(new_string("IsEmpty"))
```

```python
# Định nghĩa hàm "larger_string" nhận vào hai tham số: "text" và "n" 
def larger_string(text, n):
    # Khởi tạo một biến chuỗi rỗng tên là "result" để chứa kết quả
    result = ""

    # Sử dụng vòng lặp for để lặp lại "n" lần
    for i in range(n):
        result = result + text # Cộng dồn chuỗi "text" vào biến "result" sau mỗi vòng lặp

    # Trả về chuỗi "result" cuối cùng 
    return result

print(larger_string('abc', 2)) # Gọi hàm với đối số 'abc' và số lần lặp là 2
print(larger_string('.py', 3)) # Gọi hàm với đối số '.py' và số lần lặp là 3
```

```python
# Định nghĩa hàm list_count_4
def list_count_4(nums):
  count = 0 # Khởi tạo biến count
  # Duyệt qua từng phần tử (num) trong danh sách đầu vào (nums).
  for num in nums:
    # Kiểm tra xem phần tử hiện tại có bằng 4 hay không.
    if num == 4:
      count = count + 1 # Tăng biến đếm thêm 1.


 # Trả về tổng số lần đếm
  return count

# Gọi hàm với các danh sách khác nhau và in kết quả.
print(list_count_4([1, 4, 6, 7, 4]))  
print(list_count_4([1, 4, 6, 4, 7, 4])) 
```

```python
# Định nghĩa hàm substring_copy nhận vào hai tham số: text và n.
def substring_copy(text, n):
  flen = 2 # Đặt độ dài mặc định cho chuỗi con (substring) là 2 ký tự.
  # Kiểm tra nếu độ dài chuỗi con mong muốn lớn hơn độ dài chuỗi nhập vào.
  if flen > len(text):
    # Nếu chuỗi ngắn hơn 2, ta chỉ lấy đúng số ký tự hiện có.
    flen = len(text)

  # Trích xuất chuỗi con từ đầu chuỗi đến vị trí flen.
  substr = text[:flen]

  # Khởi tạo một chuỗi rỗng tên là 'result' để chứa kết quả cộng dồn.
  result = ""

  # Sử dụng vòng lặp để lặp lại việc cộng chuỗi con vào 'result' n lần.
  for i in range(n):
    result = result + substr

  # Trả về kết quả cuối cùng.
  return result

# Gọi hàm với các ví dụ cụ thể:
print(substring_copy('abcdef', 2)) 
print(substring_copy('p', 3))      
```
