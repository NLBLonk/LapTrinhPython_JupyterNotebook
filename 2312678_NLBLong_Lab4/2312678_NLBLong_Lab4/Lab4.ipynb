```python
plt.figure(figsize=(15, 10))

# 1. CreditScore
plt.subplot(2, 2, 1)
plt.hist(dt['CreditScore'], bins=30, color='skyblue', edgecolor='black')
plt.title('Histogram: Credit Score')
plt.xlabel('Điểm tín dụng')
plt.ylabel('Số lượng khách hàng')

# 2. Age
plt.subplot(2, 2, 2)
plt.hist(dt['Age'], bins=20, color='salmon', edgecolor='black')
plt.title('Histogram: Age')
plt.xlabel('Tuổi')

# 3. Balance
plt.subplot(2, 2, 3)
plt.hist(dt['Balance'], bins=30, color='lightgreen', edgecolor='black')
plt.title('Histogram: Balance')
plt.xlabel('Số dư tài khoản')
plt.ylabel('Số lượng khách hàng')

# 4. EstimatedSalary
plt.subplot(2, 2, 4)
plt.hist(dt['EstimatedSalary'], bins=30, color='gold', edgecolor='black')
plt.title('Histogram: Estimated Salary')
plt.xlabel('Lương ước tính')

plt.show()
```

```python
plt.figure(figsize=(10, 6))

sns.boxplot(x='Geography', y='Balance', data=dt, palette = 'Set2', hue='Geography')

plt.title('Mối quan hệ giữa Quốc gia và Số dư tài khoản')
plt.show()
```

```python
plt.figure(figsize=(10, 6))

# Vẽ boxplot cho mức lương theo quốc gia
sns.boxplot(x='Geography', y='EstimatedSalary', data=dt, palette='Set2', hue='Geography')

# Thêm tiêu đề và nhãn rõ ràng
plt.title('Mối quan hệ giữa Quốc gia và Lương ước tính', fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(10, 6))

# Vẽ boxplot cho độ tuổi theo quốc gia
sns.boxplot(x='Geography', y='Age', data=dt, palette='Set2', hue='Geography')

# Thêm tiêu đề và nhãn
plt.title('Mối quan hệ giữa Quốc gia và Độ tuổi khách hàng', fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(10, 6))

# Vẽ biểu đồ cột chồng để so sánh giới tính giữa các quốc gia
sns.countplot(x='Geography', hue='Gender', data=dt)

# Thêm tiêu đề và nhãn
plt.title('Phân bố Giới tính theo Quốc gia', fontsize=15)
plt.xlabel('Quốc gia')
plt.ylabel('Số lượng khách hàng')
plt.legend(title='Giới tính')

plt.show()
```

```python
plt.figure(figsize=(8, 6))

# Vẽ boxplot cho mức lương theo giới tính
sns.boxplot(x='Gender', y='EstimatedSalary', data=dt, palette='Set2',  hue='Gender')

# Thêm tiêu đề và nhãn
plt.title('Mối quan hệ giữa Giới tính và Lương ước tính', fontsize=15)
plt.xlabel('Giới tính')
plt.ylabel('Lương ước tính (Estimated Salary)')

plt.show()
```

```python
num_features = ['CreditScore', 'Age', 'Balance', 'EstimatedSalary']

# Thiết lập khung hình 2x2
plt.figure(figsize=(16, 12))

for i, col in enumerate(num_features):
    plt.subplot(2, 2, i+1)
    sns.boxplot(x='Exited', y=col, data=dt, hue='Exited', palette='Set2', legend=False)
    plt.title(f'Mối quan hệ: Exited vs {col}', fontsize=14)
    plt.xticks([0, 1], ['Ở lại', 'Rời đi'])

plt.show()
```

```python
cat_features = ['Geography', 'Gender', 'NumOfProducts', 'IsActiveMember', 'Complain']
plt.figure(figsize=(18, 12))

for i, col in enumerate(cat_features):
    plt.subplot(2, 3, i+1)
    sns.countplot(x=col, hue='Exited', data=dt, palette='viridis')
    plt.title(f'Mối quan hệ: {col} vs Exited')
    plt.legend(title='Exited', labels=['Ở lại', 'Rời đi'])

plt.tight_layout()
plt.show()
```

```python
cat_features = ['CreditScore', 'Age', 'Balance', 'EstimatedSalary']

plt.figure(figsize=(16, 12))

for i, col in enumerate(cat_features):
    plt.subplot(2, 2, i+1)
    sns.boxplot(x='Exited', y=col, data=dt, hue='Exited', palette='coolwarm', legend=False)
    
    plt.title(f'Mối quan hệ giữa Exited và {col}', fontsize=14)
    plt.xticks([0, 1], ['Ở lại (0)', 'Rời đi (1)'])

plt.tight_layout()
plt.show()
```
